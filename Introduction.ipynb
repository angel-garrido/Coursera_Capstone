

# 1. Introduction



### In this project the data of Seattle traffic accidents will be explored in order to discover the variables with the strongest influence to traffic collisions.

According to statista.com (https://www.statista.com/topics/3708/road-accidents-in-the-us/) 12 million vehicles were involved in crashes in the United States in 2018. Reducing traffic accidents is both an important public safety challenge and critical to prevent traffic congestions. 

Recognition of most important influences to traffic accidents might be a substantial key for improving traffic safety. Indeed, prediction of traffic accident will be crucial to improve public transportation, to design cost-effectively transportation infrastructure and to enable safer routes.

Drivers, local Seattle government, police, rescue groups and car insurance companies will gain a great insight with this analysis to make insightful decisions for reducing the number of accidents and injuries for the city. Moreover, this can be extrapolated to other cities and countries.
